### Python List

- Anteriormente cuando utilizamos split() sobre un string obtuvimos como resultado: \\
['Cuando', 'despertó,', 'el', 'dinosaurio', 'todavía', 'estaba', 'allí']

- Este objeto no es otra cosa que una secuencia de elementos. El tipo mas frecuente de secuencia de elementos son las listas.

- Las listas permiten combinar elementos de diferente tipo y son mutables. Al poder cambiar su contenido, brindan flexibilidad a la hora de manipular objetos.

- Para crear una lista utilizamos corchetes y separamos los elementos por coma.



```python
lista_1 = ['Python', 'R', 'Julia']
lista_2 = [1,2,3,4,5]
lista_3 = [1, "Pedro", 3.4]
lista_4 = ["Juan", [6, 4, 8], ['z']]
```

```python
print(lista_1[0])

print(lista_3[-2])

print(lista_4[0][1])

print(lista_4[2][0])

print(lista_4[1][2])
```

```python
lista_5 = ['A','M','I','A','J','O','V','E','N']

print(lista_5[2:5])

print(lista_5[5:])

print(lista_5[3::2]) # que vaya desde el 3° elemento hasta el final, y por último que vaya de 2 en 2.

print(lista_5[::3]) # todo, que vaya de 3 en 3

print(lista_5[:])

del lista_5[4:]
print(lista_5)
```

```python
pares=[1,3,5,7]

pares[0] = 2

print(pares)
```

```python
pares[1:4] = [4, 6, 8]

print(pares)
```

```python
pares_plus=[10,12,14,16,18,20]
pares.extend(pares_plus)
print(pares)

```

```python
pares.reverse()

print(pares)

pares.sort()

print(pares)
```

```python
lista_6 = ['uno', 'dos', 3,4]

print(lista_6 + ['item'])

print(lista_6)

print(lista_6*3)

print(lista_6)

lista_6.append('item')

print(lista_6)
```

### Python Tuple

- Las tuplas también son un objeto secuencial. Pero a diferencia de las listas, no se pueden mutar sus elementos ni añadirle nuevos. 

- Para crear una tupla utilizamos paréntesis y separamos los elementos con coma.


```python
t = (5,'Programa', 1+3j)

print(type(t))

print("t[1] = ", t[1])

print("t[0:3] = ", t[0:6])

print(len(t))
```

```python
t[0] = 10

print(t)
```

### Python Dictionary

- Clave => Valor

- El diccionario es otro objeto muy frecuentemente utilizado en Python, y junto con las listas, podrí­amos decir que son los más importantes. 

- Los diccionarios son objetos que permiten asociar claves a valores. 

- Entonces su í­ndice son las claves. Son mutables, es decir, su contenido puede ser modificado después de ser creados.Para crearlos utilizaremos llaves y los elementos se separarán por comas.



```python
d = {'key1':'valor_1','key2':'valor_2'} 
type(d)
```

```python

print("d[key1] = ", d['key1'])

print("d[key2] = ", d['key2'])
```

```python
d['key2']
```

```python
diccionario_2 = {'key1': 1,'key2': 2, 'key3':['item_0', 'item_1', 'item_2']}
```

```python
print(diccionario_2['key3'])

print(diccionario_2['key3'][0])

print(diccionario_2['key3'][0].upper())
```

```python
diccionario_2['key4'] = 'item4'

print(diccionario_2)
```

```python
print(diccionario_2.keys())

print(diccionario_2.values())

print(diccionario_2.items())
```

```python
del diccionario_2["key4"]
print(diccionario_2)
```

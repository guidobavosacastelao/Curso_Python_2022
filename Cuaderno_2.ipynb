Python Sets





- Un conjunto es una colección desordenada de elementos. Cada elemento del conjunto es único (sin duplicados) y debe ser inmutable (no se puede cambiar).

- Sin embargo, un conjunto en sí mismo es mutable. Podemos añadir o quitar elementos de la misma.

- Los conjuntos también se pueden usar para realizar operaciones matemáticas de conjuntos como unión, intersección, diferencia simétrica, etc.

````python
mi_conjunto = {1, 2, 3}
print(mi_conjunto)

mi_conjunto = {1.0, "Hola", (1, 2, 3)}
print(mi_conjunto)
````

````python
mi_conjunto = {1, 2, 3, 4, 3, 2} # no hay repetición
print(mi_conjunto)


mi_conjunto = set([1, 2, 3, 2]) # podemos generar un conjunto desde una lista
print(mi_conjunto)
````

````python
mi_conjunto = {1, 2, [3, 4]} # no puede tener elementos mutables
````

````python
mi_conjunto = {1, 3}
print(mi_conjunto)
````

````python
print(0 not in mi_conjunto)

print(3 in mi_conjunto)
````

````python
print(mi_conjunto[0])
````

````python
mi_conjunto = {1, 3}
print(mi_conjunto)

mi_conjunto.add(2) # agrego un elemento
print(mi_conjunto)

mi_conjunto.update([2, 3, 4]) # agrego varios elementos
print(mi_conjunto)

mi_conjunto.update([4, 5], {1, 6, 8}) # agrego una lista de 2 elementos y un set de 3
print(mi_conjunto)
````

````python
mi_conjunto = {1, 3, 4, 5, 6}
print(mi_conjunto)

mi_conjunto.discard(4)
print(mi_conjunto)

mi_conjunto.remove(6)
print(mi_conjunto)

mi_conjunto.discard(2) # descarto un elemento que no se encuentra en mi conjunto
print(mi_conjunto)
````

````python
mi_conjunto.remove(2) # retirar un elemento que no se encuentra en mi conjunto
print(mi_conjunto)
````

````python
mi_conjunto = set("HolaChicos")
print(mi_conjunto)

print(len(mi_conjunto)) 

print(mi_conjunto.pop()) # elimina un elemento random del conjunto

mi_conjunto.pop() # elimina otro elemento random
print(mi_conjunto) 
print(len(mi_conjunto)) 

mi_conjunto.clear() # limpio mi conjunto
print(mi_conjunto)

````

# Control de Flujo

- Los programas siguen un "flujo" definido por el programador. Hasta ahora el flujo del programa siempre fue uno sólo, se ejecuta una instrucción, luego la siguiente, luego la siguiente a esta, etc. 

- Sin embargo es posible realizar tareas diferentes dependiendo de ciertas condiciones, repetir segmentos de código, y modificar el flujo del programa de forma que no sea simplemente una secuencia lineal de instrucciones. Esto permite realizar tareas mucho más complejas, en la cuál el programa tome decisiones respecto a lo que debe hacer en función de la situación.

- Notá a continuación que todas estas estructuras terminan con un ":" , y la siguiente línea tiene un tab de indentación (como una sangría).


Estructura IF



```python
if (condicion):
    #Lo que se hace si se cumple la condición
```





```python
if (condicion):
    #Lo que se hace si se cumple la condición
else:
    #Lo que se hace si no se cumple la condición
```





```python
if (condicion1):
    #Lo que se hace si se cumple la condición1
elif (condicion2):
    #Lo que se hace si no se cumple la condicion1, pero sí la condicion2
elif (condicion3):
    #Lo que se hace si no se cumple la condicion1 ni la condición2, pero sí la condición3
else:
    #Lo que se hace si no se cumple ninguna condición
```




Ejemplos


````python
persona = "Juan"
if (persona == "Juan"):
    print("Encontrado")
````

````python
if (persona == "Pedro"):
    print("Encontrado")
else:
    print("No encontrado")
````

````python
if (persona == "Pedro"):
    print("Encontrado Pedro")
elif (persona == "Juan"): 
    print("Encontrado Juan")
else:
    print("No encontrado")
````

````python
persona = input("A quien encontraste?")

if (persona == 'Lucas'):
   print('No es Juan')

elif(persona=='Juan'):
   print('Lo encontraste')

else:
   print('Seguī buscando')
````

Cual será el resultado del siguiente código?

````python
x = 5
y = 6

if x == y:
    print(x)
elif y < 5 or y > 7:
    print(x + y)
else:
    print(y)
````

````python
x = int(input("Ingresá el valor de tu primer número"))
y = int(input("Ingresá el valor de tu segundo número"))


if x > y:
    print("Tu primer número es mayor al segundo")
elif x < y:
    print("Tu primer número es menor que el segundo")
elif x == y:
    print("Tus dos números son iguales!")
else:
    print("En este escenario se rompe todo :D")
````
